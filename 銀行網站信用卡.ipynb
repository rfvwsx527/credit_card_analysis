```python
import requests
from bs4 import BeautifulSoup

def scrape_fubon():
    url = "https://www.fubon.com/banking/Personal/credit_card/index.htm"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
    
    resp = requests.get(url, headers=headers, timeout=15)
    resp.encoding = "utf-8"
    soup = BeautifulSoup(resp.text, "lxml")
    
    cards = []
    # 每張卡片通常在 .card-item 或 .product-item 容器內
    for card in soup.select(".card-item, .card-box, .product-item"):
        cards.append({
            "bank":      "台北富邦",
            "card_name": card.select_one(".card-name, h3, .title").get_text(strip=True) if card.select_one(".card-name, h3, .title") else "",
            "reward":    card.select_one(".reward, .benefit, .desc").get_text(strip=True) if card.select_one(".reward, .benefit, .desc") else "",
            "card_url":  "https://www.fubon.com" + card.select_one("a")["href"] if card.select_one("a") else "",
        })
    return cards
```

```python
scrape_fubon()
```
