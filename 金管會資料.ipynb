```python
import requests
import pandas as pd
import io

# ✅ 正確的下載網址（直接抓 CSV）
CSV_URL = "https://data.gov.tw/dataset/45209"

# 實際 CSV 下載連結（從頁面取得）
# 進去網頁後點「CSV 檢視資料」旁的下載按鈕，取得直連網址如：
DIRECT_URL = "https://quality.data.gov.tw/dq_download_csv.php?nid=45209&md5_url=xxxx"

def download_credit_card_stats() -> pd.DataFrame:
    resp = requests.get(DIRECT_URL, timeout=30)
    resp.encoding = "utf-8-sig"
    df = pd.read_csv(io.StringIO(resp.text))
    return df
```
